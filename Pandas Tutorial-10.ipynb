# Pandas fillna()

```python
import pandas as pd
```

```python
df = pd.read_csv('Fortune_10_Handling_Missing_Values.csv')
df
```

```python
df.fillna(0)
```

```python
df.fillna(2)
```

```python
df.fillna({'Industry' : 'none' ,'Growth' : 0 })
```

```python
df.ffill()
```

```python
df.bfill()
```

```python
df.fillna(df.ffill(), axis = 0 )
```

```python
df.fillna(df.ffill(), axis = 1 )
```

```python
df.fillna(df.bfill(), axis = 0 )
```

```python
df.fillna(df.bfill(), axis = 1 )
```

```python
df.fillna(0, limit = 1 )
```

```python
df.fillna(0, limit = 2 )
```

```python
df.fillna(df.ffill(), limit = 2 )
```

```python
df.fillna(df.ffill(), limit = 1 )
```

```python
df.fillna(5,inplace = True)
df
```
