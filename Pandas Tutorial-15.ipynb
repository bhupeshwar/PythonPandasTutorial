# Pandas GroupBy Function

```python
import pandas as pd
```

```python
df = pd.read_csv('student_results_groupby.csv')
df
```

```python
gr1 = df.groupby(by = 'Section')
gr1
```

```python
gr1.groups
```

```python
df.groupby(['Section','Class']).groups
```

```python
for Section, df_1 in gr1:
    print(Section)
    print(df_1)
```

```python
list(gr1)
```

```python
dict(list(gr1))
```

```python
gr3 = df.groupby('Class').get_group(10)
gr3
```

```python
gr4 = df.groupby('Section').get_group('A')
gr4
```
