# Pandas Concat() Function #Part 2

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'ID':[1,2,3,4],
             'Name': ['A','B','C','D'],
             'Class': [5,6,7,8]})
df1
```

```python
df2 = pd.DataFrame({'ID':[3,4],
             'Name': ['C','D'],
             'Class': [7,8]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2] , axis = 1)
```

```python
pd.concat([df1,df2] , axis = 1, join = 'inner')
```

```python
df3 = pd.DataFrame({'ID':[1,2,3,4],
             'Name': ['A','B','C','D'],
             'Class': [5,6,7,8]})
df3
```

```python
df4 = pd.DataFrame({'ID':[3,4],
             'Name': ['C','D'],
             'Class': [7,8]})
df4
```

```python
pd.concat([df3,df4] , keys = ['First Data Frame','Second Data Frame'])
```

```python
pd.concat([df3,df4] , axis = 1, keys = ['First Data Frame','Second Data Frame'])
```

```python
df5 = pd.DataFrame({'ID':[1,2,3,4],
             'Name': ['A','B','C','D'],
             'Class': [5,6,7,8]})
df5
```

```python
df6 = pd.DataFrame({'Marks':[40,63,91,34]})
df6
```

```python
pd.concat([df5,df6])
```

```python
pd.concat([df5,df6] , sort = True)
```
